```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('players_fifa23.csv')
```

```python
df = df[df.BestPosition != 'GK'] # drop goalkeepers
```

```python
y = np.array(df['ValueEUR'])
```

```python
to_be_dropped = ['Name', 'FullName', 'PhotoUrl', 'Growth', 'TotalStats', 'ValueEUR','ClubNumber', 
                 'NationalNumber', 'GKDiving', 'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes',]
```

```python
df = df.drop(to_be_dropped, axis=1)
df
```

```python
top_5_nats = ['Brazil','Belgium','Argentina','France','England']

nats_from_5_to_10 = ['Italy','Spain','Netherlands','Portugal','Denmark']

nats_from_10_to_20 = ['Germany','Croatia','Mexico','Uruguay','Switzerland',
                     'United States','Colombia','Senegal','Wales','Iran',]
```

```python
df.loc[df['Nationality'].isin(top_5_nats), 'Nationality'] = 3
df.loc[df['Nationality'].isin(nats_from_5_to_10), 'Nationality'] = 2
df.loc[df['Nationality'].isin(nats_from_10_to_20), 'Nationality'] = 1
df.loc[df['Nationality'].isin([1,2,3]) == False, 'Nationality'] = 0
```

```python
df
```

```python
positions = ['GK','RB','CB','LB','RWB','CDM','LWB','RM','CM','LM','CAM','RW','RF','CF','LF','LW','ST']
df[positions] = 0
```

```python
for pos in positions:
    df.loc[df['BestPosition'] == pos, pos] += 1
    df.loc[df['ClubPosition'] == pos, pos] += 1
    df.loc[df['NationalPosition'] == pos, pos] += 1
```

```python
df.reset_index(inplace = True)

df = df.drop('index', axis=1)
```

```python
for i in range(len(df)):
    df['Positions'][i] = df['Positions'][i].split(',')
```

```python
for i in range(len(df)):
    for pos in positions:
        if pos in df['Positions'][i]:
            df[pos][i] += 1
```

```python
df
```

```python
df = df.drop(['Positions','BestPosition','ClubPosition','NationalPosition'], axis=1)
```

```python
df.loc[df['NationalTeam'].isin(top_5_nats), 'NationalTeam'] = 3
df.loc[df['NationalTeam'].isin(nats_from_5_to_10), 'NationalTeam'] = 2
df.loc[df['NationalTeam'].isin(nats_from_10_to_20), 'NationalTeam'] = 1

df.loc[df['NationalTeam'] == 'Not in team', 'NationalTeam'] = -1

df.loc[df['NationalTeam'].isin([1,2,3,-1]) == False, 'NationalTeam'] = 0
```

# EVALUATION

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV
```

```python
DF = df.copy()
```

```python
df['ground_truth'] = y
```

```python
df = df.sort_values('ground_truth')
```

```python
X = df.values
X
```

```python
X = df.values
y_Normalized = (y-y.mean())/y.std()
X_Normalized = (X-X.mean())/X.std()
```

```python
R = Ridge(alpha=1.0)
cross_val = RepeatedKFold(n_splits=10, n_repeats=5, random_state=1)
```

```python
r2 = RidgeCV(alphas=[0.001, 0.01, 0.1, 1, 10]).fit(X_Normalized, y_Normalized)
```

```python
mse = cross_val_score(R, X_Normalized, y_Normalized, cv=cross_val, scoring='neg_mean_squared_error')
print(f"MSE: {(np.mean(-mse)):.3f}")

mae = cross_val_score(R, X_Normalized, y_Normalized, cv=cross_val, scoring='neg_mean_absolute_error')
print(f"MAE: {(np.mean(-mae)):.3f}")

R2 = r2.score(X_Normalized, y_Normalized)
print(f"R^2: {(np.mean(R2)):.3f}")
```
